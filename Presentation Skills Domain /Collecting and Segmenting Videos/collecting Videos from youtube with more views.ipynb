```python
import re
import locale

locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')

searchTerms=['Presentation skill','Presentation skill delivery','Presentation skill structure','Presentation skill visual aids']
from youtubesearchpython import VideosSearch
videosList=[] 
#for s in searchTerms:
videosSearch = VideosSearch('Presentation skill', limit = 1000)


videoDetails=[]
videosCollection=[]
viewValue=False
for i in list(videosSearch.result().values()):
    for j in list(i):
        print(j)
        print()
        for k in list(j.values()):
            
            videoDetails.append(k)
            print(k)
            print()
            if k:
                if(list(k)[0])=='title':
                    for m in k.values():


                        #print(m.rpartition('views')[0])
                        try:
                            #print((m.split('seconds'))[1].split('views')[0])
                            views=int(locale.atof(((m.split('seconds'))[1].split('views')[0])))
                            if (int(views/100000)) >0:
                                viewValue=True
                                print('views = ', views)
                            else:
                                viewValue=False
                                #print('less than million views')
                        except:
                            print('no views')


                        
        if viewValue==True:
            if j not in videosList:
                
                videosList.append(j)
        viewValue=False
        videosCollection.append(videoDetails)
        videoDetails=[]

        print('*****************************')
```

```python

import xlsxwriter
workbook = xlsxwriter.Workbook('PresentationSkill Videos With Views Over 100K.xlsx')

print('# of videos = ',len(videosCollection))
for i in videosCollection:
    row=1
    col=0
    try:
        worksheet = workbook.add_worksheet(i[1])
    except:
        print('video collected')
    worksheet.write('A1', 'VideoId')
    worksheet.write('B1', 'Title')
    worksheet.write('C1', 'Popularity')
    worksheet.write('D1', 'Duration')
    worksheet.write('E1', 'Views')
    worksheet.write('F1', 'Summary')
    worksheet.write('G1', 'ChannelName')
    worksheet.write('H1', 'URL')
    
    worksheet.write(row,col,i[1])
    worksheet.write(row,col+1,i[2])
    worksheet.write(row,col+2,i[3])
    worksheet.write(row,col+3,i[4])
    worksheet.write(row,col+4,i[5]['text'])
    col+=5
    print('i[8]= ',i[8])
    print()
    oldrow=row
    if i[8]:
        for j in list(i[8]):
            worksheet.write(row,col,j['text'])
            row+=1
    row= oldrow
    worksheet.write(row,col+1,i[9]['name'])
    worksheet.write(row,col+2,i[11])
    



    
workbook.close()
```
