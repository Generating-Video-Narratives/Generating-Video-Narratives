```python
def convertTime(value):
    integer= int(value//1)
    
    decimal=str(value/1).split('.')[1]
    
    if len(decimal) ==1 and int(decimal) >0:
        decimal=int(decimal)*10
        
    finalNum=(int(integer)*60)+int(decimal)
    return finalNum

print(convertTime(0.59))
```
